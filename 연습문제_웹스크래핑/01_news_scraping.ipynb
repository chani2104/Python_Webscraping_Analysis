웹스크래핑 연습문제1 질문1

```python
import requests
from bs4 import BeautifulSoup

# 다음 경제 뉴스
url = "https://news.daum.net/economy"
print(url)

# 요청 (사람처럼 보이게)
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=headers)

# utf-8 인코딩 (중요)
res.encoding = 'utf-8'

print(type(res))        # 응답 객체 (Response)의 자료형 확인
print(res.status_code)  # 서버 요청 응답 코드 확인 (200 = 정상)


if res.ok:
    # HTML 파싱
    soup = BeautifulSoup(res.text, 'html.parser')

    # li 태그들 가져오기
    li_tags = soup.select('ul.list_newsheadline2 li')
    print(type(li_tags), len(li_tags))   # 개수 확인

    # 반복문
    for li_tag in li_tags:
        a_tag = li_tag.find('a')
        link = a_tag['href']

        # 제목
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text.strip()

        print(link)
        print(title)

else: 
    # 응답(response)이 Error 이면 status code 출력    
    print(f'Error Code = {res.status_code}')
```

웹스크래핑 연습문제1 질문2

```python
section_dict = {
    '기후/환경':'climate',
    '사회':'society',
    '경제':'economy',
    '정치':'politics',
    '국제':'world',
    '문화':'culture',
    '생활':'life',
    'IT/과학':'tech',
    '인물':'people'
}

def print_news(section_name):
    # section_dict에서 카테고리 추출
    section = section_dict[section_name]

    # URL 생성
    url = f"https://news.daum.net/{section}"
    print(f"======> {url} {section_name} 뉴스 <======")
    
    req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.3'
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if res.ok:
        
        soup = BeautifulSoup(res.text, 'html.parser')

        # li 태그들 가져오기
        li_tags = soup.select('ul.list_newsheadline2 li')

    # 반복문
        for li_tag in li_tags:
            a_tag = li_tag.find('a')
            link = a_tag['href']

            # 제목
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            title = strong_tag.text.strip()

            print(link)
            print(title)
    else:
        # 응답(response)이 Error 이면 status code 출력    
        print(f'Error Code = {res.status_code}')
```

```python
print_news('경제')
```

```python
print_news('사회')
```
